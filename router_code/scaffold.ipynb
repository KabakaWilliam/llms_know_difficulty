```python
import pandas as pd
from collections import Counter
```

```python
def majority_vote(solutions):
    texts = [sol["score"] for sol in solutions]
    return Counter(texts).most_common(1)[0][0]
```

```python
MODEL_LIST = [
    "Qwen/Qwen2.5-Math-1.5B-Instruct",
    "Qwen/Qwen2.5-Math-7B-Instruct",
    "Qwen/Qwen2.5-14B-Instruct",
    # "Qwen/Qwen2.5-Math-72B-Instruct",
]
```

```python
GREEDY_MODELS = [
    "Qwen/Qwen2.5-Math-1.5B-Instruct",
    "Qwen/Qwen2.5-Math-7B-Instruct",
    "Qwen/Qwen2.5-72B-Instruct",
]

GREEDY_MODEL_ALIASES = ["-".join(GM.split("/")) for GM in GREEDY_MODELS]
GREEDY_MODEL_ALIASES
```

```python
# TEST_PATH = "../will_replication/DATA/SR_DATA/MATH/test-Qwen-Qwen2.5-1.5B-Instruct_maxlen_3000_k_1_temp_0.0.parquet"
GEN_STR = f"maxlen_3000_k_50_temp_0.6"
TEST_PATH_STRUCT = "../will_replication/DATA/SR_DATA/MATH/test-{MODEL_ALIAS}_{GEN_STR}.parquet"
```

```python
store = {}
for GM in GREEDY_MODEL_ALIASES:
    try:
        df = pd.read_parquet(TEST_PATH_STRUCT.format(MODEL_ALIAS=GM, GEN_STR=GEN_STR))
        Pass1 = df["success_rate"].mean()
        df["majority_vote"] = df["generated_solutions"].apply(majority_vote)
        store[GM] = {"Mean SR": Pass1, "majority vote": df["majority_vote"].mean()}
    except:
        print(f"Skipping {GM}")

store #k=50, t=1
```
